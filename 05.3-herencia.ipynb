Implementar las clases `Square`, `Rectangle` y  `Circle`, sobreescribiendo los métodos `area` y `perimeter` para que muestren los resultados correctos

```python
class Square(Shape):
    """A square object."""
    
    def __init__(self, side):
        self.side = side
        
    def area(self):
        area = self.side*self.side
        return str(area)
    
    def perimeter(self):
        perimeter = 4*self.side
        return str(perimeter)
    
    def __str__(self):
        if isinstance(self, Square): 
            return "Side of Square is " + str(self.side)
    
    def __add__(self, other):
        if isinstance(self, Square):
            return Square(self.side + other.side)
        else:
            return "Shape object is not a Square"
        
    def __sub__(self, other):
        if isinstance(self, Square):
            return Square(self.side - other.side)
        else:
            return "Shape object is not a Square"
        
    def __mul__(self, other):
        if isinstance(self, Square):
            return Square(self.side * other.side)
        else:
            return "Shape object is not a Square"
        
class Circle(Shape):
    """A circle object."""
    
    def __init__(self, radious):
        self.radious = radious
        super(Circle, self).__init__()
        
    def area(self):
        area = math.pi*self.radious**2
        return str(area)
    
    def perimeter(self):
        perimeter = 2*math.pi*self.radious
        return str(perimeter)
    
    def __str__(self):
        if isinstance(self, Circle): 
            return "Side of Circle is " + str(self.radious)
    
    def __add__(self, other):
        if isinstance(self, Circle):
            return Circle(self.radious + other.radious)
        else:
            return "Shape object is not a Circle"
        
    def __sub__(self, other):
        if isinstance(self, Circle):
            return Circle(self.radious - other.radious)
        else:
            return "Shape object is not a Circle"
        
    def __mul__(self, other):
        if isinstance(self, Circle):
            return Circle(self.radious * other.radious)
        else:
            return "Shape object is not a Circle"
        
class Rectangule(Shape):
    """A rectangule object."""
    
    def __init__(self, base, height):
        self.base = base
        self.height = height
        super(Rectangule, self).__init__()
        
    def area(self):
        area = self.base * self.height
        return str(area)
    
    def perimeter(self):
        perimeter = 2*self.base + 2*self.height
        return str(perimeter)
    
    def __str__(self):
        if isinstance(self, Rectangule): 
            return "Base of Rectangule is " + str(self.base) + " and Height of Rectangule is " + str(self.height)
    
    def __add__(self, other):
        if isinstance(self, Rectangule):
            return Rectangule(self.base + other.base, self.height + other.height)
        else:
            return "Shape object is not a Rectangule"
        
    def __sub__(self, other):
        if isinstance(self, Rectangule):
            return Rectangule(self.base - other.base, self.height - other.height)
        else:
            return "Shape object is not a Rectangule"
        
    def __mul__(self, other):
        if isinstance(self, Rectangule):
            return Rectangule(self.base * other.base, self.height * other.height)
        else:
            return "Shape object is not a Rectangule"
```
